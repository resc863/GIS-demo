```python
import numpy as np
import pandas as pd
```

```python
filename = "/content/저층주거 침수피해 시각화_20221201.csv"
df = pd.read_csv(filename, encoding="utf-8")
df.head()
```

```python
longitude = []
latitude = []

for i in df['공간정보']:
    longitude.append(i.split()[1][1:])
    latitude.append(i.split()[2][:-1])
```

```python
for i in range(len(longitude)):
    print(longitude[i], latitude[i])
```

```python
df["longitude"] = longitude
df["latitude"] = latitude
```
